### Αλγόριθμος παραγωγής λαβυρίνθων
```
1. Αρχικά όλα τα κελιά περιέχουν εμπόδια και η λίστα των τοίχων είναι κενή.
2. Ένα κελί επιλέγεται ως μέρος του λαβυρίνθου. Οι τοίχοι του κελιού προστίθενται στη λίστα των τοίχων.
3. Όσο η λίστα των τοίχων δεν είναι κενή:
      1. Επιλέγεται τυχαία ένας τοίχος από τη λίστα. Εαν μόνο ένα από τα (δύο) κελιά που χωρίζει ο τοίχος είναι μέρος του λαβυρίνθου, τότε:
            1. Ο τοίχος και το κελί που δεν είναι μέρος του λαβυρίνθου, γίνονται μέρος του λαβυρίνθου.
            2. Οι τοίχοι των νέων κελιών που προστέθηκαν στο λαβύρινθο προστίθενται στη λίστα των τοίχων.
      2. Ο τοίχος αφαιρείται από τη λίστα των τοίχων
```
*ως τοίχοι ενός κελιού θεωρούνται όσα απο τα 4 γειτονικά κελιά (όχι τα διαγώνια) περιέχουν εμπόδιο

````
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from IPython.display import HTML

class visualization:
    def __init__(self, S, F):
        '''
          Η μέθοδος αυτή αρχικοποιεί ένα αντικείμενο τύπου visualization.
          Είσοδος: 
          -> S: το σημείο εκκίνσης της αναζήτησης
          -> F: το σημείο τερματισμού
        '''
        self.S = S
        self.F = F
        self.images = []
    
    def draw_step(self, grid, frontier, expanded_nodes):
        '''
          Η συνάρτηση αυτή καλείται για να σχεδιαστεί ένα frame στο animation (πρακτικά έπειτα από την επέκταση κάθε κόμβου)
          Είσοδος: 
          -> grid: Ένα χάρτης τύπου grid
          -> frontier: Μια λίστα με τους κόμβους που ανήκουν στο μέτωπο της αναζήτησης
          -> expanded_nodes: Μια λίστα με τους κόμβους που έχουν ήδη επεκταθεί
          Επιστρέφει: None
          Η συνάρτηση αυτή πρέπει να καλεστεί τουλάχιστον μια φορά για να μπορέσει να σχεδιαστει ένα animation (πρεπεί το animation να έχει τουλάχιστον ένα frame).
        '''
        image = np.zeros((grid.N, grid.N, 3), dtype=int)
        image[grid.grid == 0] = [255, 255, 255]
        image[grid.grid == 1] = [0, 0, 0]
        
        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        self.images.append(image)
    
    def add_path(self, path):
        '''
          Η συνάρτηση αυτή προσθέτει στο τελευταίο frame το βέλτιστο μονοπάτι.
          Είσοδος:
          -> path: Μια λίστα η όποια περιέχει το βέλτιστο μονοπάτι (η οποία πρέπει να περιέχει και τον κόμβο αρχή και τον κόμβο στόχο)
          Έξοδος: None
        '''
        for n in path[1:-1]:
            image = np.copy(self.images[-1])
            image[n] = [66, 221, 245]
            self.images.append(image)
        for _ in range (100):
            self.images.append(image)
        
    def create_gif(self, fps = 30, repeat_delay = 2000):
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        fig = plt.figure()
        plt.axis('off')
        ims = []
        for img in self.images:
            img = plt.imshow(img)
            ims.append([img])
        ani = animation.ArtistAnimation(fig, ims, interval=1000//fps, blit=True, repeat_delay= repeat_delay)
        plt.close(fig)
        return ani
        
    def save_gif(self, filename, fps = 30):
        '''
            Η συνάρτηση αυτή ξαναδημιουργεί και αποθηκεύει το animation σε ένα αρχείο.
            Είσοδος:
            -> Το όνομα του αρχείου με κατάληξη .gif
            Έξοδος: (None)
        '''
        ani = self.create_gif(fps)
        writer = PillowWriter(fps= fps)
        ani.save(filename, writer=writer)

    def show_gif(self, fps= 30, repeat_delay = 2000):
        '''
            Η συνάρτηση αυτή εμφανίζει inline το animation.
            Είσοδος:
            -> fps: τα frames per second
            Έξοδος: Το αντικείμενο που παίζει το animation
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        ani = self.create_gif(fps, repeat_delay)
        # return HTML(ani.to_html5_video())
        return HTML(ani.to_jshtml())

    def show_last_frame(self):
        '''
            Η μέθοδος αυτή εμφανίζει inline το τελευταίο frame που έχει δημιουργήθει.
            Είσοδος:
            Έξοδος: Το αντικείμενο που εμφανίζει την εικόνα.
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        else:
            plt.imshow(self.images[-1])


class EmptyStackOfImages(Exception):
    pass
````

````
# from visualization import *
def func(self):
  return self.f 
#Η κλάση node χρησιμοποιείται για να κρατούνται οι πληροφορίες που αφορούν τους κόμβους που επισκέπτεται ο αλγόριθμος
class node:
      def __init__(self,position,parent):
        self.position = position
        self.parent = parent
        self.g = 0 # Distance to start node
        self.f = 0 # Total cost with heuristic 
      def __eq__(self, other):
        return self.position == other.position
            
class pathfinder:
    def __init__(self, S, F, grid, c, h):
        self.S = S
        self.F = F
        self.grid = grid
        self.cost = c
        self.heuristic = h
        #for visualization 
        self.vis = visualization(S, F)
        self.path=[] 
      
    
    

    def find_path(self):
        ## YOUR CODE HERE
        goal = self.F
        start = self.S
        S = node(self.S,None)
        F = node(self.F,None)
        open_list = []
        #Η open_list έχει τα στοιχεία μετάβασης
        open_list.append(S)
        #Η closed_list έχει τους κόμβους που έχουν επεκταθεί
        closed_list = []
        while open_list:
          current_node = open_list.pop(0)
          #Αν το κελί που εξετάζεται τώρα είναι ο τελικός στόχος τότε η διαδικασία πρέπει να βρεθεί το path, ακολουθώντας τους πατέρες των κόμβων μέχρι
          #να φτάσουμε στο S και έπειτα η διαδικασία πρέπει να τελειώσει 
          if current_node.position == goal:
            while current_node.position != start:
              self.path.append(current_node.position)
              current_node = current_node.parent
            self.path.append(start)
            break
          (x,y) = current_node.position
          #Αν το κελί δεν είναι ο στόχος τότε πρέπει να βρεθούν τα παιδία του, δηλαδή τα 4 γειτονικά του κελιά
          #από τα παιδία του αποδεκτά είναι αυτά που βρίσκονται εντός ορίων του πίνακα και που δεν είναι τοίχος.
          #Για τα αποδεκτά κελιά εξετάζεται αρχικά αν βρίσκονται στην open_list και αν χρειάζεται αν ανανεωθούν οι τιμές g, f και parent
          #Αλλιώς εξετάζεται αν βρίσκονται στην closed_list και αν χρειάζεται να ανανεωθούν οι ίδιες τιμές με πριν.
          #Τέλος αν τα κελιά αυτά επισκέπτονται για πρώτη φορά πρέπει αν γίνουν νέοι κόμβοι και να προστεθούν στην open_list
          kids = [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]
          for next_move in kids:
            if next_move[0] < 0 or next_move[0] > self.grid.N-1 or next_move[1] < 0 or next_move[1] > self.grid.N-1:
              continue
            state = self.grid.grid[next_move]
            if state == 1:
              continue
            kid = node(next_move,current_node)
            if kid in open_list:
              new_g = current_node.g + self.cost(kid.position[0],kid.position[1],goal[0],goal[1])
              if new_g < kid.g:
                kid.f = new_g +kid.f - kid.g
                kid.g = new_g
                kid.parent = current_node
            elif kid in closed_list:
              new_g = current_node.g + self.cost(kid.position[0],kid.position[1],goal[0],goal[1])
              if new_g < kid.g:
                kid.f = new_g +kid.f - kid.g
                kid.g = new_g
                kid.parent = current_node
            else:
              kid.g = current_node.g + self.cost(kid.position[0],kid.position[1],goal[0],goal[1])
              kid.f = kid.g + self.heuristic(kid.position[0],kid.position[1],goal[0],goal[1])
              open_list.append(kid)
          closed_list.append(current_node)
          #η open_list ταξινομείται κάθε φορά ως προς την τιμή f των κόμβων της
          open_list.sort(key = func)
        pass
        
    def get_path(self):
      self.find_path()
      return self.path

````

**Απαντήσεις των ερωτήσεων του μέρους 2:**

Όταν η τιμή g(n) είναι ίση με 0 και υπάρχει ευριστική συνάρτηση, είτε αυτή είναι η manhattan είτε η ευκλείδια, τότε προκύπτει ο αλγόριθμος αναζήτησης best first. Σε αυτήν την περίπτωση δεν αναμένεται να βρεθεί η βέλτιστη λύση.

Όταν η τιμή g(n) είναι 1 και η ευριστική συνάρτηση h είναι 0, τότε προκύπτει ο αλγόριθμος Dijkstra, ο οποίος βρίσκει πάντα την βέλτιστη λύση.

Τέλος, όταν η g(n) έχει την τιμή 1 και η ευριστική συνάρτηση h είναι διάφορη του 0, είτε είναι η ευκλείδια είτε η manhattan, τότε προκύπτει ο αλγόριθμος A*. Σε αυτή την περίπτωση, δεδομένου του ότι όλες οι ευριστικές δίνουν τιμές μικρότερες ή ίσες των πραγματικών αποστάσεων, ο αλγόριθμος αυτός αναμένεται να βρίσκει πάντα το βέλτιστο μονοπάτι.

Μία επιπλέον ευριστική που προτείνεται είναι αυτή της διαγώνιας απόστασης μεταξύ ενός ελεύθερου κελιού και του τελικού στόχου.

````
#Mέρος 2 τροποποιημένο
#Η πρώτη αλλαγή που έχει γίνει είναι ότι πλέον όταν υπολογίζονται τα g του κάθε κελιού λαμβάνονται υπόψιν οι τιμές του κελιού που υπάρχουν στον πίνακα 
#self.grid.costs
#H δεύτερη αλλαγή είναι ότι έχει προστεθεί μία μεταβλητή closed_nodes η οποία όταν τελειώσει η διαδικασία εύρεσης μονοπατιού θα περιέχει την τιμή των κόμβων 
#που επεκτάθηκαν κατά τη διάρκειά της. Η τιμή της μεταβλητής αυτής επιστρέφεται από τη συνάρτηση get_closed_nodes


# from visualization import *
def func(self):
  return self.f 

class node:
      def __init__(self,position,parent):
        self.position = position
        self.parent = parent
        self.g = 0 # Distance to start node
        self.f = 0 # Total cost with heuristic 
      def __eq__(self, other):
        return self.position == other.position
            
class pathfinder:
    def __init__(self, S, F, grid, h):
        self.S = S
        self.F = F
        self.grid = grid
        self.heuristic = h
        #for visualization 
        self.vis = visualization(S, F)
        self.closed_nodes = 0
        self.path=[] 
      
    
    

    def find_path(self):
        ## YOUR CODE HERE
        goal = self.F
        start = self.S
        S = node(self.S,None)
        F = node(self.F,None)
        open_list = []
        open_list.append(S)
        closed_list = []
        while open_list:
          current_node = open_list.pop(0)
          if current_node.position == goal:
            while current_node.position != start:
              self.path.append(current_node.position)
              current_node = current_node.parent
            self.path.append(start)
            self.closed_nodes = len(closed_list)
            break
          (x,y) = current_node.position
          kids = [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]
          for next_move in kids:
            if next_move[0] < 0 or next_move[0] > self.grid.N-1 or next_move[1] < 0 or next_move[1] > self.grid.N-1:
              continue
            state = self.grid.grid[next_move]
            if state == 1:
              continue
            kid = node(next_move,current_node)
            if kid in open_list:
              new_g = current_node.g + self.grid.costs[(kid.position[0],kid.position[1])]
              if new_g < kid.g:
                kid.f = new_g +kid.f - kid.g
                kid.g = new_g
                kid.parent = current_node
            elif kid in closed_list:
              new_g = current_node.g + self.grid.costs[(kid.position[0],kid.position[1])]
              if new_g < kid.g:
                kid.f = new_g +kid.f - kid.g
                kid.g = new_g
                kid.parent = current_node
            else:
              kid.g = current_node.g + self.grid.costs[(kid.position[0],kid.position[1])]
              kid.f = kid.g + self.heuristic(kid.position[0],kid.position[1],goal[0],goal[1])
              open_list.append(kid)
          closed_list.append(current_node)
          open_list.sort(key = func)
        pass
        
    def get_path(self):
      self.find_path()
      return self.path
    def get_closed_nodes(self):
      return self.closed_nodes
````

**Απαντήσεις των ερωτήσεων του Μέρους 3:**

Για τη λειτουργία των αλγορίθμων προκύπτουν τα εξής συμπεράσματα:

Αρχίκα, αξίζει να σημειωθέι ότι ο λαβύρινθος δεν έχει πολλά μονοπάτια από την αρχή μέχρι το τέλος και για αυτό τις περισσότερες φορές το μήκος των μονοπατιών ταυτίζεται. Πορ'όλα αυτά, όταν δημιουργούνται περισσότερα του ενός μονοπάτια, οι αλγόριθμοι συνήθως εμφανίζουν μεταξύ τους διαφορετικές λύσεις. Αυτό γίνεται επειδή οι ευριστικές δίνουν μεγαλύτερες τιμές από τις πραγματικές αποστάσεις των κόμβων από το τέλος. Επιπλέον, παρατηρείται ότι αν μία ευριστική δίνει μεγαλύτερες τιμές από τις πραγματικές και μία δεύτερη ακόμα μεγαλύτερες, δεν σημαίνει ότι η πρώτη θα δώσει απαραίτητα συντομότερο μονοπάτι από την δεύτερη.

Αν το τυχαίο κόστος μετάβασης έπαιρνε τιμές στο διάστημα (1,2), τότε ο Α* θα έβρισκε βέλτιστο μονοπάτι με όλες τις ευριστικές αλλά θα καθυστερούσε περισσότερο απ'ότι στην περίπτωση που το κόστος μετάβασης ήταν 1. Αυτό γίνεται επειδή τώρα οι τιμές των ευριστικών είναι πιο υποτιμημένες.

Τέλος, αν το κόστος μετάβασης έπαιρνε τιμές στο διάστημα [0,0.1), τότε καμία από τις ευριστικές που χρησιμοποιούνται δεν θα έβρισκε πάντα το βέλτιστο μονοπάτι, εκτός φυσικά από την περίπτωση που h = 0 και άρα ο Α* εκφυλίζεται σε Dijkstra. Τα αποτελέσματα σε αυτή την περίπτωση θα ήταν παρόμοια με αυτά που παρατηρήθηκαν στην άσκηση.
